```python
import pandas as pd

df = pd.read_csv("Data/zomato_data.csv")
df.head()
```

```python
df.info()
```

```python
df["rate"] = df["rate"].str.replace("/5", "", regex=False)
df["rate"] = pd.to_numeric(df["rate"])
```

```python
print("Average Rating:", round(df["rate"].mean(), 2))
```

```python
print("Average Cost:", round(df["approx_cost(for two people)"].mean(), 2))
```

```python
df["listed_in(type)"].value_counts()
```

```python
df["online_order"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df["listed_in(type)"].value_counts().plot(kind="bar")

plt.title("Restaurant Types")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df["online_order"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Online Orders Distribution")
plt.ylabel("")
plt.show()
```

```python
sns.histplot(df["rate"], bins=10)

plt.title("Ratings Distribution")
plt.show()
```

```python
df["listed_in(type)"].value_counts().plot(kind="bar")

plt.title("Restaurant Types")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```
